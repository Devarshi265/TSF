## Exploratory data analysis
### ploting different features to plot dataset

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=sample1)
plt.title('Category vs Sub Category')
plt.xlabel('Sub-Catgory')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.heatmap(storedata.corr(),cmap='Purples',annot=True)
```

```python
sample1.hist(bins=50 ,figsize=(20,15))
```

```python
figsize=(15,10)
sns.pairplot(sample1,hue='Sub-Category')
plt.show
```

```python
# state wise count of sample data
plt.figure(figsize=(18,5))
sns.countplot(storedata['State'])
plt.xticks(rotation=90)
plt.title('State',fontsize=20)
```

```python
plt.figure(figsize=(10,4))
sns.lineplot('Discount','Profit',data=sample1 , color='y',label='Discount')
plt.legend()
plt.show()
```

# region wise analysis



```python
sample1_region = sample1.groupby(['Region'])[['Sales', 'Discount', 'Profit']].mean()
sample1_region
```

```python
sample1_region.plot.pie(subplots=True, 
                   figsize=(18, 20), 
                   autopct='%1.1f%%',
                   labels = sample1_region.index)
```

## ship mode wise analysis

```python
storedata['Ship Mode'].value_counts()
```

```python
storedata_shipmode = storedata.groupby(['Ship Mode'])[['Sales', 'Discount', 'Profit']].mean()
storedata_shipmode.plot.pie(subplots=True,
                     figsize=(18, 20), 
                     autopct='%1.1f%%', 
                     labels = storedata_shipmode.index)

#sales are higher in same day
#discount is higher on first class
# profit is higher on first class
```
